```python
import csv
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

# data source https://people.sc.fsu.edu/~jburkardt/data/csv/csv.html
#header, day_of_week, crashes_on_day = [], [], []
with open ("data/crash_catalonia.csv") as f:
    reader = csv.reader(f)
    # header of csv, I could use these for axis titles later
    header = next(reader)
    print(header)
    day_of_week, crashes_on_day = [], []
    for row in reader:
        #print(row)
        # row[0] = day of the week
        try:
            day = row[0]
            num_crashes = int(row[1])
        # For some foresaken reason, mine is reading in two blank lines.  
        #   Would not have caught without printing row and seeing it
        except:
            print("No good data here!")
        else: 
            day_of_week.append(day)
            #row[1] = number of crashes on that day 
            crashes_on_day.append(num_crashes)
# let's verify we got stuff
print(day_of_week)
print(crashes_on_day)

```

```python
# what does this get?
#   how many days havehad x threshold of crashes - y is crash threshold, x is how many days met that threshold
plt.hist(crashes_on_day)

# we have 7 days in a week - 7 bins?
num_bins = 7
plt.hist(crashes_on_day, num_bins)

# histograms are focused on bins... what if we used a bar chart?
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(day_of_week, crashes_on_day)

#fig.show()
#plt.title('Height Distribution of US Presidents')
#plt.xlabel('height (cm)')
#plt.ylabel('number');
```
